```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ruta = '/home/sofia/t1-algoco/Tarea1-Inf221/ordenamiento/salida_selection_sort2.txt' 
df = pd.read_csv(ruta, sep = ' ', header = None, names=['Tamaño', 'Tiempo'])
df['Tiempo'] 
df = df[df['Tamaño'] <= 1e6]
df.sort_values('Tamaño', inplace=True)
```

```python
plt.figure(figsize=(20, 12))  # Ajusta el tamaño del gráfico según necesites
plt.scatter(df['Tamaño'], df['Tiempo'], marker='o', linestyle='-', color='hotpink')
plt.title('Tiempo de ejecución vs Tamaño del arreglo')
plt.xlabel('Tamaño del arreglo')
plt.ylabel('Tiempo de ejecución (nanosegundos)')
plt.grid(True)
plt.xscale('log')  # Solo escala logarítmica en x
plt.yscale('log')  # Escala lineal en y
plt.show()
```
