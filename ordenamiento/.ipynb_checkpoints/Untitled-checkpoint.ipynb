```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ruta_ins = 'salidas para graficar/salida_insertion_sort2.txt'
df = pd.read_csv(ruta_ins, sep=' ', header=None, names=['Tama単o', 'Tiempo'])
df['Tama単o'] = df['Tama単o'].astype(int)
df['Tiempo'] = df['Tiempo'].astype(float)
df.sort_values(by='Tama単o', inplace=True)
```

```python
plt.figure(figsize=(15, 9))
plt.scatter(df['Tama単o'], df['Tiempo'], label='Insertion Sort', color='deeppink', alpha = 0.5)
plt.xlabel('Tama単o del arreglo')
plt.ylabel('Tiempo [ns]')
plt.title('Insertion Sort')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.show()
```

```python
ruta_merge = 'salidas para graficar/salida_merge_sort2.txt'
df = pd.read_csv(ruta_merge, sep=' ', header=None, names=['Tama単o', 'Tiempo'])
df['Tama単o'] = df['Tama単o'].astype(int)
df['Tiempo'] = df['Tiempo'].astype(float)
df.sort_values(by='Tama単o', inplace=True)
```

```python
plt.figure(figsize=(15, 9))
plt.scatter(df['Tama単o'], df['Tiempo'], label='Merge Sort', color='deeppink', alpha = 0.5)
plt.xlabel('Tama単o del arreglo')
plt.ylabel('Tiempo [ns]')
plt.title('Merge Sort')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.show()
```
