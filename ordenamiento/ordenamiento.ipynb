```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ruta_ins_al = 'salidas para graficar/ins_sort_salida_al.txt'
ruta_ins_inv = 'salidas para graficar/ins_sort_salida_inv.txt'
ruta_ins_ord = 'salidas para graficar/ins_sort_salida_ord.txt'
```

```python
df_ins_al = pd.read_csv(ruta_ins_al, sep=" ", header=None, names=['Tamaño', 'Tiempo'])
df_ins_inv = pd.read_csv(ruta_ins_inv, sep=" ", header=None, names=['Tamaño', 'Tiempo'])
df_ins_ord = pd.read_csv(ruta_ins_ord, sep=" ", header=None, names=['Tamaño', 'Tiempo'])

df_ins_al['Tamaño'] = df_ins_al['Tamaño'].astype(int)
df_ins_inv['Tamaño'] = df_ins_inv['Tamaño'].astype(int)
df_ins_ord['Tamaño'] = df_ins_ord['Tamaño'].astype(int)
df_ins_al['Tiempo'] = df_ins_al['Tiempo'].astype(float)
df_ins_inv['Tiempo'] = df_ins_inv['Tiempo'].astype(float)
df_ins_ord['Tiempo'] = df_ins_ord['Tiempo'].astype(float)

df_ins_al.sort_values(by=['Tamaño'], inplace=True)
df_ins_inv.sort_values(by=['Tamaño'], inplace=True)
df_ins_ord.sort_values(by=['Tamaño'], inplace=True)
```

```python
ruta_merge_al = 'salidas para graficar/merge_sort_salida_al.txt'
ruta_merge_inv = 'salidas para graficar/merge_sort_salida_inv.txt'
ruta_merge_ord = 'salidas para graficar/merge_sort_salida_ord.txt'
```

```python
df_merge_al = pd.read_csv(ruta_merge_al, sep=" ", header=None, names=['Tamaño', 'Tiempo'])
df_merge_inv = pd.read_csv(ruta_merge_inv, sep=" ", header=None, names=['Tamaño', 'Tiempo'])
df_merge_ord = pd.read_csv(ruta_merge_ord, sep=" ", header=None, names=['Tamaño', 'Tiempo'])

df_merge_al['Tamaño'] = df_merge_al['Tamaño'].astype(int)
df_merge_inv['Tamaño'] = df_merge_inv['Tamaño'].astype(int)
df_merge_ord['Tamaño'] = df_merge_ord['Tamaño'].astype(int)
df_merge_al['Tiempo'] = df_merge_al['Tiempo'].astype(float)
df_merge_inv['Tiempo'] = df_merge_inv['Tiempo'].astype(float)
df_merge_ord['Tiempo'] = df_merge_ord['Tiempo'].astype(float)

df_merge_al.sort_values(by=['Tamaño'], inplace=True)
df_merge_inv.sort_values(by=['Tamaño'], inplace=True)
df_merge_ord.sort_values(by=['Tamaño'], inplace=True)
```

```python
ruta_quick_al = 'salidas para graficar/quick_sort_salida_al.txt'
ruta_quick_inv = 'salidas para graficar/quick_sort_salida_inv.txt'
ruta_quick_ord = 'salidas para graficar/quick_sort_salida_ord.txt'
```

```python
ruta_introsort_al = 'salidas para graficar/introsort_salida_al.txt'
ruta_introsort_inv = 'salidas para graficar/introsort_salida_inv.txt'
ruta_introsort_ord = 'salidas para graficar/introsort_salida_ord.txt'
```

```python
df_introsort_al = pd.read_csv(ruta_introsort_al, sep=" ", header=None, names=['Tamaño', 'Tiempo'])
df_introsort_inv = pd.read_csv(ruta_introsort_inv, sep=" ", header=None, names=['Tamaño', 'Tiempo'])
df_introsort_ord = pd.read_csv(ruta_introsort_ord, sep=" ", header=None, names=['Tamaño', 'Tiempo'])

df_introsort_al['Tamaño'] = df_introsort_al['Tamaño'].astype(int)
df_introsort_inv['Tamaño'] = df_introsort_inv['Tamaño'].astype(int)
df_introsort_ord['Tamaño'] = df_introsort_ord['Tamaño'].astype(int)
df_introsort_al['Tiempo'] = df_introsort_al['Tiempo'].astype(float)
df_introsort_inv['Tiempo'] = df_introsort_inv['Tiempo'].astype(float)
df_introsort_ord['Tiempo'] = df_introsort_ord['Tiempo'].astype(float)

df_introsort_al.sort_values(by=['Tamaño'], inplace=True)
df_introsort_inv.sort_values(by=['Tamaño'], inplace=True)
df_introsort_ord.sort_values(by=['Tamaño'], inplace=True)
```

```python
plt.figure(figsize=(15, 9))
plt.scatter(df_ins_al['Tamaño'], df_ins_al['Tiempo'], label='Insertion sort (arreglo aleatorio)', color='deeppink', alpha=0.5)
plt.scatter(df_merge_al['Tamaño'], df_merge_al['Tiempo'], label='Merge sort (arreglo aleatorio)', color='dodgerblue', alpha=0.5)
plt.scatter(df_quick_al['Tamaño'], df_quick_al['Tiempo'], label='Quick sort (arreglo aleatorio)', color='gold', alpha=0.5)
plt.scatter(df_introsort_al['Tamaño'], df_introsort_al['Tiempo'], label='Introsort (arreglo aleatorio)', color='orchid', alpha=0.5)
plt.xlabel('Tamaño del arreglo')
plt.ylabel('Tiempo [ns]')
plt.title('Comparación de algoritmos')
plt.grid()
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```
